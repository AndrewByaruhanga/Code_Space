```python
from tkinter import *
from PIL import Image, ImageTk
from datetime import date
from tkinter import StringVar
from tkinter import filedialog
from tkinter import messagebox
from tkcalendar import DateEntry
import os
from tkinter.ttk import Combobox
import openpyxl,xlrd
from openpyxl import Workbook
import pathlib


def validate_fields():
    if not Name.get() or not NIN.get():
        messagebox.showerror("Error", "NIN/FCS and NAME are mandatory fields.")
        return False
    return True

background = "#A52A2A"

root = Tk() 
root.title("Company name")
root.geometry("1200x700+100+3")
root.resizable(False, False) 
root.configure(bg=background)

file = pathlib.Path('Andy_Portfolio.csv')
if file.exists():
    pass
else:
    file = Workbook()
    sheet = file.active
    sheet['A1'] = "NIN/FCS"
    sheet['B1'] = "Loan Application Date"
    sheet['C1'] = "Name"
    sheet['D1'] = "Date of Birth"
    sheet['E1'] = "Marital_status"
    sheet['F1'] = "Residence Description"
    sheet['G1'] = "Education level"
    sheet['H1'] = "Number of dependants"
    sheet['I1'] = "Number of years in service"
    sheet['J1'] = "Monthly Income"
    sheet['K1'] = "Loan Amount requested"
    sheet['L1'] = "Loan Term Requested"
    sheet['M1'] = "Loan Type Requested"
    sheet['N1'] = "Number of CRB Inquiries"
    sheet['O1'] = "Number of Current/Active loans"
    sheet['P1'] = "Number of previous loans fully paid"
    sheet['Q1'] = "Number of previous loans written off"
    sheet['R1'] = "Credit Mix"
    sheet['S1'] = "Days in arrears for current facilities"
    sheet['T1'] = "Days in arrears for previous facilities"
    sheet['U1'] = "Credit Risk Classification"
    file.save('Company.csv')


# Top
Label(root, text="Powered by: Andy", width=10, height=1, bg="#A52A2A", anchor='e').pack(side=TOP, fill=X)
Label(root, text="Andy ", width=10, height=1, bg="#A52A2A", fg="white", font="Arial 20 bold").pack(side=TOP, fill=X)


framebg = "#EDEDED"
framefg = "#1f77b4"

#National Identification number/Financial card number
Label(root,text="NIN/FCS:",font = "arial 13",fg= framebg, bg=background).place(x=200, y=65)
NIN = StringVar()
NIN_entry = Entry(root, textvariable=NIN, width=15,font="arial 13")
NIN_entry.place(x=370,y=67)

#Loan Application Date
Label(root,text="LOAN APPLICATION DATE:",font = "arial 13",fg= framebg, bg=background).place(x=700, y=65)
Date = StringVar()
today = date.today()
curr_date = today.strftime("%d/%m/%Y")
date_entry = Entry(root,textvariable=Date,width=15,font="arial 13")
date_entry.place(x=923,y=65)
Date.set(curr_date)

#Static Details
obj = LabelFrame(root,text = "Static Details",font=20,bd=2,width=1145,bg=framebg,fg=framefg,height=230,relief=GROOVE)
obj.place(x=20,y=115)
Label(obj,text="Name:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=15)
Label(obj,text="Gender:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=45)
Label(obj,text="Date of Birth:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=75)
Label(obj,text="Marital status:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=105)
Label(obj,text="Residence Description:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=135)
Label(obj,text="Product Type:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=165)


Label(obj,text="Number of dependants:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=15)
Label(obj,text="Number of years in Boda Business:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=45)
Label(obj,text="Source of repayment:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=75)
Label(obj,text="Region:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=105)
Label(obj,text="Education level:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=135)	
Label(obj,text="Purchase Type:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=165)

#Full_name
Name=StringVar()
name_entry = Entry(obj,textvariable=Name,width=20,font = 'arial 10')
name_entry.place(x=350,y=15)

#Gender section    
gen=Combobox(obj,values=["Male","Female"],font="Roboto 10",width=17,state="r")
gen.place(x=350,y=45)

gender_weights = {
    "Male": 0,
    "Female": 0
}

def selectiongender():
    gender_status = gen.get()
    gender_status_weight = gender_weights.get(gender_status, 0)
    return gender_status_weight

#Date of birth and age Calculator

def calculate_age(event):
    birth_date = DOB_entry.get_date()
    today = date.today()
    Age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return Age

DOB_entry = DateEntry(obj, width=20, background='#1f77b4', foreground='white', borderwidth=2,date_pattern='dd/mm/yyyy')
DOB_entry.place(x=350, y=75)

#Classifying age
def classify_age(n):
    if n < 18:
        return "Younger_than_18"
    elif 18 <= n < 25:
        return "18 to 25 yrs old"
    elif 25 <= n < 30:
        return "25 to 30 yrs old"
    elif 30 <= n < 40:
        return "30 to 40 yrs old"
    elif 40 <= n < 55:
        return "40 to 55 yrs old"
    else:
        return "Older_than_55"

def update_age_classification(event):
    age = calculate_age(event)
    classification = classify_age(age)
    Age_classification.set(classification)

DOB_entry.bind("<<DateEntrySelected>>", update_age_classification)
Age_classification = StringVar()

#Age Classification Window

age_weights = {
    "Younger_than_18": 0,
    "18 to 25 yrs old": 3.5,
    "25 to 30 yrs old":4.5,
    "30 to 40 yrs old":5,
    "40 to 55 yrs old":3.5,
    "Older_than_55":1.5
}

def selectionage():
    age_classification = Age_classification.get()
    age_weight = age_weights.get(age_classification, 0)
    return age_weight
    
#Marital status
marital_status_weights = {
    "Single": 2,
    "Married": 3,
    "Divorced": 3,
    "Widow/Widower": 2
}

def selectionmarital():
    marital_status = Marital.get()
    marital_status_weight = marital_status_weights.get(marital_status, 0)
    return marital_status_weight
        
Marital=Combobox(obj,values=["Single","Married","Divorced","Widow/Widower"],font="Roboto 10",width=17,state="r")
Marital.place(x=350,y=105)

#Residence
residence_weights = {
    "Owned": 4,
    "Rented": 2,
    "Mortgaged": 3,
    "Other": 1
}
def selectionresidence():
    residence_status = Resi.get()
    residence_weight = residence_weights.get(residence_status, 0)
    return residence_weight

Resi = Combobox(obj, values=["Owned", "Rented", "Mortgaged", "Other"], font="Roboto 10", width=17, state="r")
Resi.place(x=350, y=135)

#Product type
product_weights = {
    "New Boda": 0,
    "Used Boda": 0
}
def selectionproduct():
    product_status = prod.get()
    product_weight = product_weights.get(product_status, 0)
    return product_weight

prod = Combobox(obj, values=["New Boda", "Used Boda"], font="Roboto 10", width=17, state="r")
prod.place(x=350, y=165)


#Dependants
dependants_weights = {
    "0": 4,
    "1": 3,
    "2": 3,
    "3": 2,
    "4": 2,
    "5": 1,
    "6": 1,
    "7": 0,
    "8": 0,
    "9": 0,
    ">=10": 0
}
def selectiondependants():
    dependants = Depend.get()
    dependants_weight = dependants_weights.get(dependants, 0)
    return dependants_weight

Depend = Combobox(obj, values=["0", "1", "3", "4","5","6","7","8","9",">=10"], font="Roboto 10", width=17, state="r")
Depend.place(x=900, y=15)

#Years in service
years_service_weights = {
    "<2 years": 1,
    "2 to 5 years": 2,
    "6 to 10 years": 3,
    "11 to 15 years": 4,
    ">15 years": 5
}
def selectionyearsservice():
    years_service = Num_years_service.get()
    years_service_weight = years_service_weights.get(years_service, 0)
    return years_service_weight


Num_years_service = Combobox(obj, values=["<2 years", "2 to 5 years", "6 to 10 years", "11 to 15 years", ">15 years"], font="Roboto 10", width=17, state="r")
Num_years_service.place(x=900, y=45)

#Source of repayment
Source_of_repayment_weights = {
    "Riding Boda":2,
    "Salary":3,
    "Both Salary & Riding Boda":4
}
def selectionsourcerepayment():
    repayment = Source_of_repayment.get()
    source_repayment_weight = Source_of_repayment_weights.get(repayment, 0)
    return source_repayment_weight
    
Source_of_repayment_value = StringVar()
Source_of_repayment = Combobox(obj,values=["Riding Boda","Salary","Both Salary & Riding Boda"],font="Roboto 10",width=17,state="r", textvariable=Source_of_repayment_value)
Source_of_repayment.place(x=900,y=75)


#Region
Region_value = StringVar()
Region = Combobox(obj,values=["Buganda_Region","Rwenzori_Region","Ankole_Region","Acholi Sub_Region","West_Nile_Region",
                                    "Teso Sub_Region","Karamoja Sub_Region","Busoga Sub_Region","Mount Elgon Sub_Region","Kigezi Sub_Region"],font="Roboto 10",width=17,state="r", textvariable=Region_value)
Region.place(x=900,y=105)


Regional_weights = {
    "Buganda_Region": 2,
    "Rwenzori_Region": 3,
    'Ankole_Region': 3,
    'Acholi Sub_Region':4,
    'West_Nile_Region':4,
    'Teso Sub_Region':5,
    'Karamoja Sub_Region':5,
    'Busoga Sub_Region':4,
    'Mount Elgon Sub_Region':4,
    'Kigezi Sub_Region':3
}

def selection_region():
    region_type = Region.get()
    Region_type_weight = Regional_weights.get(region_type)
    return Region_type_weight


#Education
education_weights = {
    "No Education ": 2,
    "Secondary Education": 2,
    "University Education": 3,
    "Post Graduate": 4
}
def selectioneducation():
    education_level = Educ.get()
    education_weight = education_weights.get(education_level, 0)
    return education_weight

Educ = Combobox(obj, values=["No Education", "Secondary Education", "University Education", "Post Graduate"], font="Roboto 10", width=17, state="r")
Educ.place(x=900, y=135)

#Education
purchase_weights = {
    "Ride forself purchase": 1,
    "Fleet purchase": 2
}
def selectioneducation():
    purchase_level = purch.get()
    purchase_weight = purchase_weights.get(purchase_level, 0)
    return purchase_weight

purch = Combobox(obj, values=["Ride forself purchase", "Fleet purchase"], font="Roboto 10", width=17, state="r")
purch.place(x=900, y=165)

#Financial Details
obj = LabelFrame(root,text = "Financial Details",font=20,bd=2,width=1145,bg=framebg,fg=framefg,height=190,relief=GROOVE)
obj.place(x=20,y=375)

Label(obj,text="Monthly Income:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=20)
Label(obj,text="Loan_Amount:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=50)
Label(obj,text="Tenure:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=80)
Label(obj,text="Down Payment:",font = "arial 13",bg= framebg, fg=framefg).place(x=30, y=110)
#Debt to Income Ratio - (x=30, y=135)
#Loan to Security Value Ratio - (x=30, y=165)


Label(obj,text="Number of CRB Inquiries:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=20)
Label(obj,text="Credit Mix:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=50)
Label(obj,text="Number of Current_Active Loans:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=80)
Label(obj,text="Number of Previous Loans Fully Paid:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=110)
Label(obj,text="Credit_risk_classification:",font = "arial 13",bg= framebg, fg=framefg).place(x=600, y=140)


monthIncome = IntVar()
monthIncome_entry = Entry(obj, textvariable=monthIncome, width=20, font='arial 10')
monthIncome_entry.place(x=350, y=15)

Loan_Amount=IntVar()
Loan_Amount_entry = Entry(obj,textvariable=Loan_Amount,width=20,font = 'arial 10')
Loan_Amount_entry.place(x=350,y=45)

tenure=IntVar()
tenure_entry = Combobox(obj,values=["1","2","3","6","12","18","24"],font="Roboto 10",width=17,state="r", textvariable=tenure)
tenure_entry.place(x=350,y=75)

Down_payment=Combobox(obj,values=["750,000","1,000,000","1,250,000","1,500,000","2,000,000","2,500,000","3,000,000","3,500,000","4,000,000"],font="Roboto 10",width=17,state="r")
Down_payment.place(x=350,y=105)

Down_payment_weights = {
       "750,000": 1,
       "1,000,000": 2,
        "1,250,000":3,
        "1,500,000":3.5,
        "2,000,000":4,
        "2,500,000":4.5,
        "3,000,000":5,
        "3,500,000":5.5,
        "4,000,000":6
}
def selectioncustomertype():
    downpaym = Down_payment.get()
    Down_payment_weight = Down_payment_weights.get(downpaym, 0)
    return Down_payment_weight

#DTI classification
def classify_DTI(num):
    if num >= 60:
        return "Above 60% DTI"
    elif 40 <= num < 60:
        return "40% to 60% DTI"
    else:
        return "Below 40% DTI"

def calculate_DTI():
    monthly_income_str = monthIncome_entry.get()
    loan_amount_str = Loan_Amount_entry.get()
    tenure_str = tenure_entry.get()
    
    if monthly_income_str and loan_amount_str and tenure_str:
        monthly_income = int(monthly_income_str)
        loan_amount = int(loan_amount_str)
        tenure = int(tenure_str)
        
        if tenure != 0:
            dti_ratio = ((loan_amount / tenure) * 100) / monthly_income
            dti_classification = classify_DTI(dti_ratio)
            DTI.set(dti_classification)
        else:
            DTI.set("null")
    else:
        DTI.set("null")

calculate_button = Button(obj, text="Calculate Debt to Income ratio", command=calculate_DTI, font=('arial', 13), fg=framefg)
calculate_button.place(x=30, y=140, height=20)

DTI = StringVar()
DTI_entry = Entry(obj, textvariable=DTI, width=20,font=('arial', 10, 'italic'))
DTI_entry.place(x=350, y=140)
        
DTI_weights = {
    "Above 60% DTI": 1,
    "40% to 60% DTI": 2,
    "Below 40% DTI": 3
}

def selectionDTI():
    DTI_ = DTI_entry.get()
    DTI_weight = DTI_weights.get(DTI_, 0)
    return DTI_weight

     
#Right pane

    
#Number of crb inquiries
Number_of_crb_inquiries=Combobox(obj,values=["0", "1", "2","3", "4","5",">=6"],font="Roboto 10",width=17,state="r")
Number_of_crb_inquiries.place(x=900,y=20)

crb_inquiries_weights = {
       "0": 2,
       "1": 3,
       "2": 3,
       "3": 3,
       "4": 2,
       "5": 2,
       ">=6": 1    
}
def selectioncrbinquiries():
    crb_inquiries = Number_of_crb_inquiries.get()
    crb_inquiries_weight = crb_inquiries_weights.get(crb_inquiries, 0)
    return crb_inquiries_weight



#Credit mix
Creditmix = Combobox(obj,values=["None", "2","3", ">=4"], font="Roboto 10", width=17, state="r")
Creditmix.place(x=900, y=50)

Creditmix_weights = {
       "None": 3,
       "2": 2,
       "3": 2,
       ">=4": 1    
}
def selectionCreditmix():
    Creditmixx = Creditmix.get()
    Creditmix_weight = Creditmix_weights.get(Creditmixx, 0)
    return Creditmix_weight

#Current active loans
Current_active_loans = Combobox(obj, values=["0", "1","2","3",">=4"], font="Roboto 10", width=17, state="r")
Current_active_loans.place(x=900, y=80)

Number_of_Current_Active_Loans_weights = {
       "0": 0,
       "1": 3,
       "2": 2,
       "3":1,
       ">=4": 1      
}
def selectionCurrent_Active_Loans():
    current_loans = Current_active_loans.get()
    current_Active_Loans_weight = Number_of_Current_Active_Loans_weights.get(current_loans, 0)
    return current_Active_Loans_weight


#Fully paid loans
Number_of_previous_fully_paid_loans_weights = {
       "0": 0,
       "1": 1,
       "2": 2,
       "3":3,
       ">=4": 3      
}
def selectionfullypaidloans():
    previous_loans = previous_fully_paid_loans.get()
    previous_fully_paid_loans_weight = Number_of_previous_fully_paid_loans_weights.get(previous_loans, 0)
    return previous_fully_paid_loans_weight

previous_fully_paid_loans = Combobox(obj, values=["0", "1","2","3",">=4"], font="Roboto 10", width=17, state="r")
previous_fully_paid_loans.place(x=900, y=110)

#Credit risk classification
credit_risk_weights = {
    "Standard": 5,
    "Normal": 4,
    "Watch": -1,
    "Sub-standard": -1,
    "Doutful": -1,
    "Loss": -2
}

def selectioncredit_risk():
    credit_risk = Creditriskclassification.get()
    credit_risk_weight = credit_risk_weights.get(credit_risk, 0)
    return credit_risk_weight

Creditriskclassification = Combobox(obj, values=["Standard","Normal", "Sub-standard", "Doutful", "Watch", "Loss"], font="Roboto 10", width=17, state="r")
Creditriskclassification.place(x=900, y=140)

#Decision plane


SCOS = StringVar()
AD = StringVar()

CS_entry = Entry(root,width=20,font="arial 15",justify='center')
CS_entry.place(x=760,y=610,width=80,height=20)
SCOS_entry = Entry(root, textvariable=SCOS, width=20,font="arial 15",justify='center')
SCOS_entry.place(x=760,y=630,width=80, height=20)


# gender_weights + age_weight + marital_status_weight + residence_weight + education_weight + rank_weight + dependants_weight + years_service_weight 
# + source_repayment_weight + Employer_Salary_Payment_weight + region_land_type_weight + loan_type_weight + DTI_weight + LSV_weight +
# customer_type_weight + crb_inquiries_weight + Creditmix_weight + current_Active_Loans_weight + previous_fully_paid_loans_weight + credit_risk_weight

max_weighted_value = (
    max(gender_weights.values()) +
    max(age_weights.values()) +
    max(marital_status_weights.values()) +
    max(residence_weights.values()) +
    max(education_weights.values()) +
    max(dependants_weights.values()) +
    max(years_service_weights.values()) +
    max(Source_of_repayment_weights.values()) +
    max(product_weights.values()) +
    max(purchase_weights.values()) +
    max(Down_payment_weights.values()) +
    max(DTI_weights.values()) +
    max(crb_inquiries_weights.values()) +
    max(Creditmix_weights.values()) +
    max(Number_of_Current_Active_Loans_weights.values()) +
    max(Number_of_previous_fully_paid_loans_weights.values()) +
    max(credit_risk_weights.values())
)

min_weighted_value = (
    min(gender_weights.values()) +
    min(age_weights.values()) +
    min(marital_status_weights.values()) +
    min(residence_weights.values()) +
    min(education_weights.values()) +
    min(dependants_weights.values()) +
    min(years_service_weights.values()) +
    min(Source_of_repayment_weights.values()) +
    min(product_weights.values()) +
    min(purchase_weights.values()) +
    min(Down_payment_weights.values()) +
    min(DTI_weights.values()) +
    min(crb_inquiries_weights.values()) +
    min(Creditmix_weights.values()) +
    min(Number_of_Current_Active_Loans_weights.values()) +
    min(Number_of_previous_fully_paid_loans_weights.values()) +
    min(credit_risk_weights.values())
)

def calculate_weight():
    if validate_fields():
      Customer_Score_Value = 0

    if gen.get():
       Customer_Score_Value += gender_weights.get(gen.get(), 0)
    if Age_classification.get():
       Customer_Score_Value += age_weights.get(Age_classification.get(), 0)
    if Marital.get():
       Customer_Score_Value += marital_status_weights.get(Marital.get(), 0)
    if Resi.get():
       Customer_Score_Value += residence_weights.get(Resi.get(), 0)
    if Educ.get():
        Customer_Score_Value += education_weights.get(Educ.get(), 0)
    if Depend.get():
        Customer_Score_Value += dependants_weights.get(Depend.get(), 0)
    if Num_years_service.get():
        Customer_Score_Value += years_service_weights.get(Num_years_service.get(), 0)
    if Source_of_repayment.get():
        Customer_Score_Value += Source_of_repayment_weights.get(Source_of_repayment.get(), 0)
    if prod.get():
        Customer_Score_Value += product_weights.get(prod.get(), 0)
    if purch.get():
        Customer_Score_Value += purchase_weights.get(purch.get(), 0)
    if Down_payment.get():
        Customer_Score_Value += Down_payment_weights.get(Down_payment.get(), 0)        
    if DTI_entry.get():
        Customer_Score_Value += DTI_weights.get(DTI_entry.get(), 0) 


    if Number_of_crb_inquiries.get():
        Customer_Score_Value += crb_inquiries_weights.get(Number_of_crb_inquiries.get(), 0)
    if Creditmix.get():
        Customer_Score_Value += Creditmix_weights.get(Creditmix.get(), 0)
    if Current_active_loans.get():
        Customer_Score_Value += Number_of_Current_Active_Loans_weights.get(Current_active_loans.get(), 0)
    if previous_fully_paid_loans.get():
        Customer_Score_Value += Number_of_previous_fully_paid_loans_weights.get(previous_fully_paid_loans.get(), 0)
    if Creditriskclassification.get():
        Customer_Score_Value += credit_risk_weights.get(Creditriskclassification.get(), 0) 
    

    final_Customer_Score_Value = Customer_Score_Value
    CS_entry.delete(0, END) 
    CS_entry.insert(0, int(final_Customer_Score_Value))
    CS_entry.config(justify="center")
    CS_entry.config(font=("arial 10", 12))
    return final_Customer_Score_Value

calculate_button = Button(root, text="Calculate Customer Score", command=calculate_weight)
calculate_button.place(x=370, y=610, height=20)

#Define percentiles
max_value = max_weighted_value
min_value = min_weighted_value
Approval_range_value = max_value-min_value
Approval_mean_value  = (max_value-min_value)/2


#Cut off score entry
SCOS_entry.delete(0, END) 
SCOS_entry.insert(0, int(Approval_mean_value))
SCOS_entry.config(justify="center")
SCOS_entry.config(font=("arial 10", 12))

calculate_button = Button(root, text="Approval Cut-Off Score", command=Approval_mean_value)
calculate_button.place(x=370, y=630,height=20)

AD_entry = Entry(root, width=20,font="arial 15",justify='center')
AD_entry.place(x=670,y=650,width=285, height=20)

def get_approval_decision():
    if validate_fields():
        final_Customer_Score_Value = calculate_weight()
        loann_amount = int(Loan_Amount_entry.get())  
    if final_Customer_Score_Value <= Approval_mean_value:
            approval_decision = "Loan Declined"
            pass
    elif final_Customer_Score_Value > Approval_mean_value and int(Loan_Amount_entry.get()) > 15000000:
        approval_decision = "Loan referred to Decision Hierarchy"
    else:
            approval_decision = "Loan Approved"

    AD_entry.delete(0, END) 
    AD_entry.insert(0, approval_decision)
    AD_entry.config(justify="center")
    AD_entry.config(font=("arial 10", 12))     
    return approval_decision


# Create the button with the function to get the approval decision
calculate_button = Button(root, text="Approval Decision", command=get_approval_decision)
calculate_button.place(x=370, y=650, height=20)


approved_amount = StringVar()
AA_entry = Entry(root, width=20,font="arial 15",justify='center')
AA_entry.place(x=670,y=670,width=285, height=20)

def get_approved_value():
    if validate_fields():
      final_Customer_Score_Value = calculate_weight()
      loann_amount = int(Loan_Amount_entry.get())

    if final_Customer_Score_Value < Approval_mean_value:
        approved_amount = 0
    elif final_Customer_Score_Value > Approval_mean_value and int(Loan_Amount_entry.get()) > 15000000:
        approved_amount = "Pending Hierarchy Approval"
    else:
        approved_amount = loann_amount       
        
    AA_entry.delete(0, END) 
    AA_entry.insert(0, (approved_amount))
    AA_entry.config(justify="center")
    AA_entry.config(font=("arial 10", 12))     
    return approved_amount
  
# The approval decision
calculate_button = Button(root, text="Approved Loan", command=get_approved_value)
calculate_button.place(x=370, y=670, height=20)


root.mainloop()


```
